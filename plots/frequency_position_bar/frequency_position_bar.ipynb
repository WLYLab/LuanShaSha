```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(15,4), dpi=300,facecolor=(1, 1,1))#平铺画布
plt.subplots_adjust(wspace =0.35, hspace =0.3)#调整子图间距

ax = plt.subplot(121)
recipe = ["375 g A","75 g C","250 g G","300 g T"]
ingredients = [x.split()[-1] for x in recipe]
data=pd.read_csv("./S4_pos.csv",header=None,index_col=None)
N = len(data.values[0])
ind = np.arange(N)
width = 0.35
A_values=np.array(data.values[0])
C_values=np.array(data.values[1])
A_C_values=np.array(A_values+C_values)
G_values=np.array(data.values[2])
A_C_G_values=np.array(A_values+C_values+G_values)
T_values=np.array(data.values[3])
p1 = plt.bar(ind, A_values,color='#4682B4')
p2 = plt.bar(ind, C_values,color='#66CDAA',bottom=A_values)
p3 = plt.bar(ind, G_values,color='#F0E68C', bottom=A_C_values)
p3 = plt.bar(ind, T_values,color='#FAA460', bottom=A_C_G_values)
plt.xlabel('Position',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
#plt.title('Scores by group and gender')
num=[]
for i in range(-20,21):
    if i%5==0:
        num.append(i)
    else:
        num.append('')
plt.xlim(-1, 41)
plt.ylim(-0.02, 1.0)
plt.xticks(ind,num,fontsize=13)
plt.yticks(np.arange(0, 1.1, 0.5),fontsize=13)
#plt.legend(('A', 'C','G','T'))
legend=plt.legend(ingredients,
          title="Base",
          frameon=False,
          loc="center left",
          labelspacing=0.2,
          handlelength=1.5,
          handleheight=1.5,
          handletextpad=0.3,
          borderaxespad=-0.3,
          fontsize=10,
          bbox_to_anchor=(1, 0, 0.1, 1)
                 )
#legend.get_frame().set_facecolor('#00FFCC')
legend.get_title().set_fontsize(fontsize = 12)# 设置Base的大小
ax.spines['top'].set_visible(False)#去掉上面的边框
ax.spines['right'].set_visible(False)#去掉右边的边框

plt.text(-7,0.99,'A',fontsize='xx-large')


ax = plt.subplot(122)
recipe = ["375 g A","75 g C","250 g G","300 g T"]
ingredients = [x.split()[-1] for x in recipe]
data=pd.read_csv("./S4_neg.csv",header=None,index_col=None)
N = len(data.values[0])
ind = np.arange(N)
width = 0.35
A_values=np.array(data.values[0])
C_values=np.array(data.values[1])
A_C_values=np.array(A_values+C_values)
G_values=np.array(data.values[2])
A_C_G_values=np.array(A_values+C_values+G_values)
T_values=np.array(data.values[3])
p1 = plt.bar(ind, A_values,color='#4682B4')
p2 = plt.bar(ind, C_values,color='#66CDAA',bottom=A_values)
p3 = plt.bar(ind, G_values,color='#F0E68C', bottom=A_C_values)
p3 = plt.bar(ind, T_values,color='#FAA460', bottom=A_C_G_values)
plt.xlabel('Position',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
#plt.title('Scores by group and gender')
num=[]
for i in range(-20,21):
    if i%5==0:
        num.append(i)
    else:
        num.append('')
plt.xlim(-1, 41)
plt.ylim(-0.02, 1.0)
plt.xticks(ind,num,fontsize=13)
plt.yticks(np.arange(0, 1.1, 0.5),fontsize=13)
#plt.legend(('A', 'C','G','T'))
legend=plt.legend(ingredients,
          title="Base",
          frameon=False,
          loc="center left",
          labelspacing=0.2,
          handlelength=1.5,
          handleheight=1.5,
          handletextpad=0.3,
          borderaxespad=-0.3,
          fontsize=10,
          bbox_to_anchor=(1, 0, 0.1, 1))
#legend.get_frame().set_facecolor('#00FFCC')
legend.get_title().set_fontsize(fontsize = 12)# 设置Base的大小
ax.spines['top'].set_visible(False)#去掉上面的边框
ax.spines['right'].set_visible(False)#去掉右边的边框
plt.text(-7,0.99,'B',fontsize='xx-large')

#plt.show()
plt.savefig('./frequency_position_bar.pdf',bbox_inches='tight')#不能跟plt.show同时用，可用来保存svg,pdf,tiff等格式
```
