```python
# used as plot program
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(figsize=(15,6), dpi=300,facecolor=(1, 1, 1))#平铺画布，figsize用来设置长宽
# plt.tight_layout()#调整整体空白
plt.subplots_adjust(wspace =0.3, hspace =0.5)#调整子图间距


plt.subplot(231)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S1.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("C.elegans",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(A)',fontsize='xx-large') #在文中写字


plt.subplot(232)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S2.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("D.melanogaster",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(B)',fontsize='xx-large') #在文中写字

plt.subplot(233)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S3.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("A.thaliana",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(C)',fontsize='xx-large') #在文中写字


plt.subplot(234)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S4.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("E.coli",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(D)',fontsize='xx-large') #在文中写字


plt.subplot(235)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S5.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("G.subterraneus",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(E)',fontsize='xx-large') #在文中写字


plt.subplot(236)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="./data/draw_line_chart/S6.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
#x=range(len(all_results))
x= list(map(lambda x: x*10, range(len(all_results))))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index*10+1)+","+str(max_value)+"]",xytext=(max_value_index*10-0.2*length_x,max_value),xy=(max_value_index*10,max_value),fontsize=11)
plt.plot([max_value_index*10,max_value_index*10],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线

plt.plot(x,all_results)#画折线图
plt.title("G.pickeringii",fontsize=12)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x*10]) #设置x轴坐标
plt.ylim((0.5, 1))#设置  y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x*10, 100)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*10*length_x*1.0,0.98,'(F)',fontsize='xx-large') #在文中写字




#plt.savefig('./BestACC_line_chart.pdf',bbox_inches='tight') #保存图片，可以保存格式 pdf, svg,tiff。其中 bbox_inches用来是图片更加紧凑
 #print "done"
plt.show()#显示图片，不可与plt.savefig同时用
```
